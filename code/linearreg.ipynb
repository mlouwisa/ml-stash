# Linear Regression with One Variable

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris = iris[['petal_length', 'petal_width']]
iris
```

```python
X = iris['petal_length']
y = iris['petal_width']
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 23)
```

```python
X_train
```

```python
import numpy as np
X_train = np.array (X_train).reshape(-1, 1)
X_train
```

```python
X_test = np.array (X_test).reshape(-1, 1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
Y_pred_train = m*X_train + c
Y_pred_train.flatten()
```

**or use can also use predict formula in scikit**

```python
y_pred_train1 = lr.predict(X_train)
y_pred_train1
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred_train1, color='red')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

we used the train X and Y to train the model, now we try to predict for test and see if it will generate the same best fitting line.

```python
y_pred_test1 = lr.predict(X_test)
y_pred_test1
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred_test1, color='red')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

It can generalize prediction for unseen data.
